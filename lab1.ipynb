Цель: построить графики абсолютной погрешности каждого метода в зависимости от шага численного дифференцирования $h_n = \frac{2}{2^n}$ для $n=1..21$ для набора из 5 функций:
$$sin(x^2), \space\space cos(sin(x)), \space\space exp(sin(cos(x)), \space\space ln(x+3),  \space\space \sqrt{x+3}.$$ <br>

Точку $x$ выберем случайно, а сравнивать вычисленную по обычной формуле производную будем с вычисленной численным методом по 5 формулам: 
1. $$\frac{f(x+h) - f(x)}{h}$$
2. $$\frac{f(x) - f(x-h)}{h}$$
3. $$\frac{f(x+h) - f(x-h)}{h}$$
4. $$\frac{4}{3}\frac{f(x+h)-f(x-h)}{2h} - \frac{1}{3}\frac{f(x+2h)-f(x-2h)}{4h}$$
5. $$\frac{3}{2}\frac{f(x+h)-f(x-h)}{2h} -\frac{3}{5}\frac{f(x+2h)-f(x-2h)}{4h} + \frac{1}{10}\frac{f(x+3h)-f(x-3h)}{6h}$$

Результатом исследования будут графики абсолютной погрешности - разности аналитического и численного расчета производных в зависимости от размера $h$ в численном методе.6h}$

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.pi/12
n = np.arange(1, 22, 1)
h = np.array(2/(2**(n)))

#формулы для вычисления погрешностей численным методом
def der_f(x, dh, f, num):
    match num:
        case 1: 
            return (      (f(x+dh) - f(x))    / dh)
        case 2: 
            return (      (f(x)    - f(x-dh)) / dh)
        case 3: 
            return (      (f(x+dh) - f(x-dh)) / (2*dh))
        case 4: 
            return (4/3 * (f(x+dh) - f(x-dh)) / (2*dh) - 1/3 * (f(x+2*dh) - f(x-2*dh)) / (4*dh))
        case 5: 
            return (3/2 * (f(x+dh) - f(x-dh)) / (2*dh) - 3/5 * (f(x+2*dh) - f(x-2*dh)) / (4*dh) \
                                                      + 1/10 * (f(x+3*dh) - f(x-3*dh)) / (6*dh))
```

#### 1) $sin(x^2)$
1я функция $f_1 = sin(x^2)$, ее производная $f'_1 = sin(x^2)$

```python
def f(x): return np.sin(x**2)
def der_f_real(x): return 2*x*np.cos(x**2)
ans = der_f_real(x)
```

```python
plt.figure(figsize=[13, 6], dpi=350)
plt.plot(h, np.array(abs(ans - der_f(x, h, f, 1))), '.-', label='y1')
plt.plot(h, np.array(abs(ans - der_f(x, h, f, 2))), '.-', label='y2')
plt.plot(h, np.array(abs(ans - der_f(x, h, f, 3))), '.-', label='y3')
plt.plot(h, np.array(abs(ans - der_f(x, h, f, 4))), '.-', label='y4')
plt.plot(h, np.array(abs(ans - der_f(x, h, f, 5))), '.-', label='y5')
    
plt.xlabel("$h$")
plt.ylabel("$der_{real} - der(h)$")
plt.xscale('log')
plt.yscale('log')
plt.legend()
plt.title("Абсолютная погрешность методов для f1=sin(x^2)")
plt.grid()
plt.show()
```

#### 2) $cos(sin(x))$
2я функция $f_2 = cos(sin(x))$, ее производная $f'_2 = -cos(x)sin(sin(x))$

```python
def f(x): return np.cos(np.sin(x))
def der_f_real(x): return -np.cos(x)*np.sin(np.sin(x))
ans = der_f_real(x)
```

```python
plt.figure(figsize=[13, 6], dpi=350)
plt.plot(h, np.array(abs(ans - der_f(x, h, f, 1))), '.-', label='y1')
plt.plot(h, np.array(abs(ans - der_f(x, h, f, 2))), '.-', label='y2')
plt.plot(h, np.array(abs(ans - der_f(x, h, f, 3))), '.-', label='y3')
plt.plot(h, np.array(abs(ans - der_f(x, h, f, 4))), '.-', label='y4')
plt.plot(h, np.array(abs(ans - der_f(x, h, f, 5))), '.-', label='y5')

plt.xlabel("$h$")
plt.ylabel("$der_{real} - der(h)$")
plt.xscale('log')
plt.yscale('log')
plt.legend()
plt.title("Абсолютная погрешность методов для f2=cos(sin(x))")
plt.grid()
plt.show()
```

#### 3) $exp(sin(cos(x)))$
3я функция $f_3 = exp(sin(cos(x)))$, ее производная $f'_3 = -sin(x)*cos(cos(x))*exp(sin(cos(x)))$

```python
def f(x): return np.exp(np.sin(np.cos(x)))
def der_f_real(x): return -np.sin(x) * np.cos(np.cos(x)) * np.exp(np.sin(np.cos(x)))
ans = der_f_real(x)
```

```python
plt.figure(figsize=[13, 6], dpi=350)
plt.plot(h, np.array(abs(ans - der_f(x, h, f, 1))), '.-', label='y1')
plt.plot(h, np.array(abs(ans - der_f(x, h, f, 2))), '.-', label='y2')
plt.plot(h, np.array(abs(ans - der_f(x, h, f, 3))), '.-', label='y3')
plt.plot(h, np.array(abs(ans - der_f(x, h, f, 4))), '.-', label='y4')
plt.plot(h, np.array(abs(ans - der_f(x, h, f, 5))), '.-', label='y5')

plt.xlabel("$h$")
plt.ylabel("$der_{real} - der(h)$")
plt.xscale('log')
plt.yscale('log')
plt.legend()
plt.title("Абсолютная поргрешность методов для f3=exp(sin(cos(x)))")
plt.grid()
plt.show()
```

#### 4) $ln(x+3)$
4я функция $f_4 = ln(x+3)$, ее производная $f'_4 = \frac{1}{x+3}$

```python
def f(x): return np.log(x+3)
def der_f_real(x): return 1/(x+3)
ans = der_f_real(x)
```

```python
plt.figure(figsize=[13, 6], dpi=350)
plt.plot(h, np.array(abs(ans - der_f(x, h, f, 1))), '.-', label='y1')
plt.plot(h, np.array(abs(ans - der_f(x, h, f, 2))), '.-', label='y2')
plt.plot(h, np.array(abs(ans - der_f(x, h, f, 3))), '.-', label='y3')
plt.plot(h, np.array(abs(ans - der_f(x, h, f, 4))), '.-', label='y4')
plt.plot(h, np.array(abs(ans - der_f(x, h, f, 5))), '.-', label='y5')

plt.xlabel("$h$")
plt.ylabel("$der_{real} - der(h)$")
plt.xscale('log')
plt.yscale('log')
plt.legend()
plt.title("Абсолютная поргрешность методов для f4=ln(x+3)")
plt.grid()
plt.show()
```

#### 5) $\sqrt{x+3}$
5я функция $f_5= \sqrt{x+3}$, ее производная $f'_5 = \frac{1}{2\sqrt{x+3}}$

```python
def f(x): return np.sqrt(x+3)
def der_f_real(x): return 1/(2*np.sqrt(x+3))
ans = der_f_real(x)
```

```python
plt.figure(figsize=[13, 6], dpi=350)
plt.plot(h, np.array(abs(ans - der_f(x, h, f, 1))), '.-', label='y1')
plt.plot(h, np.array(abs(ans - der_f(x, h, f, 2))), '.-', label='y2')
plt.plot(h, np.array(abs(ans - der_f(x, h, f, 3))), '.-', label='y3')
plt.plot(h, np.array(abs(ans - der_f(x, h, f, 4))), '.-', label='y4')
plt.plot(h, np.array(abs(ans - der_f(x, h, f, 5))), '.-', label='y5')

plt.xlabel("$h$")
plt.ylabel("$der_{real} - der(h)$")
plt.xscale('log')
plt.yscale('log')
plt.legend()
plt.title("Абсолютная поргрешность методов для f5=sqrt(x+3)")
plt.grid()
plt.show()
```

### Вывод
По графикам можно определить, что до определенных значений абсолютной погрешности зависимости линейные. Нелинейность возникает при заходе величины погрешности в область достаточно малых чисел, появляется машинная погрешность. Можно сделать вывод, что стремление сделать шаг $h$ как можно меньше бессмысленно, поскольку, например, погрешность результата при $h = 10^{-1}$ для одного из методов примерно равна погрешности для $h = 10^{-6}$.


```python
def f(x): return np.exp(np.sin(np.cos(x)))
def der_f_real(x): return -np.sin(x) * np.cos(np.cos(x)) * np.exp(np.sin(np.cos(x)))
ans = der_f_real(x)

y3 = np.array(abs(ans - der_f(x, h, f, 3)))
y4 = np.array(abs(ans - der_f(x, h, f, 4)))
y5 = np.array(abs(ans - der_f(x, h, f, 5)))

print('h1_3', y3.argmin())
print('h1_4', y4.argmin())
print('h1_5', y5.argmin())
```
